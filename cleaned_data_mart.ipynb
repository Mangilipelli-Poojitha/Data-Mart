```python
import numpy as np
import pandas as pd
import math
```

```python
df=pd.read_csv("data_mart.csv")
df
```

```python
df.isnull().sum()
```

```python
df['segment']=df['segment'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Date']=pd.to_datetime(df.week_date)
```

```python
df.head()
```

```python
df['Day']=df["Date"].dt.day
```

```python
def calculate_week_number(day):
  if day>=1 and day<=7:
    return 1
  elif day>=8 and day<=14:
    return 2
  elif day>=15 and day<=21:
    return 3
  elif day >=22 and day<=28:
    return 4
  else:
    return 5


df['week_number'] = df['Day'].apply(calculate_week_number)
```

```python
df.head()
```

```python
df['month_number']=df['Date'].dt.month
df['calendar_year']=df['Date'].dt.year
df.tail()
```

```python
df['month_name']=df['Date'].dt.strftime('%b')
df
```

```python
def map_demographic(s):
  s=str(s)
  if s[0]=='C':
    return 'Couples'
  elif s[0]=='F':
    return 'Families'
  else:
    return 'unknown'



df['demographic'] = df['segment'].apply(map_demographic)
```

```python
df.head()
```

```python
def map_age_band(s):
  s=str(s)
  if s[1]=='1':
    return 'Young Adults'
  elif s[1]=='2':
    return 'Middle Aged'
  elif s[1]=='3' or s[1]=='4':
    return 'Retirees'
  else:
    return 'unknown'


df['age_band']=df['segment'].apply(map_age_band)
```

```python
df.head()
```

```python
df['avg_transaction']=round(df['sales']/df['transactions'],2)
df
```

```python
df.to_csv('cleaned_data_mart.csv',index=False)
```
